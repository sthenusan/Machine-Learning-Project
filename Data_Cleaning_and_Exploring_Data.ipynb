# First step in the project is the Data Exploring and Cleaning 

### Problem definition

Water is one of the most important resources for both individuals and governments. Even pure water has no value. Tanzania is a developing country with a difficulty with clean water. According to reports in worldometers.info,Tanzania has a population of 59,353,795 people, and it must provide clean water to more than 59 million people. Plenty water wells have already been established, but there are also many that are either worthless or in need of repair.

Drivendata has launched a competition called 'Pump It Up' to draw attention to this issue. People can use the prediction model to figure out whether water points are functional, nonfunctional, or functional but need to be repaired. This model can assist the Tanzanian government in identifying wells that are likely to require maintenance or provide helpful information for future wells.

The project's goal is to create a model that shows the state of water points (as functional, nonfunctional, functional but needs repair). With this model, It will show Tanzanian authorities how to make productive use of water resources. It also aids the government's investment in wells.

## Data Source

The original data can be downloaded by the [DrivenData 'Pump it Up: Data Mining the Water Table' competition](https://www.drivendata.org/competitions/7/pump-it-up-data-mining-the-water-table/page/23/). There are four distinct data sets: submission format, training set, test set, and train labels set, which comprises well status. Competitors are expected to develop a predictive model using the given training set and labels, then apply it to the test set to determine the status of the wells and submit it.

Train set and train label set will be used in this project. There are 59400 water locations in the train set, with 40 features. The train labels data contains 59400 identical water points with the same train set, but only contains information about the id and status of these points.
